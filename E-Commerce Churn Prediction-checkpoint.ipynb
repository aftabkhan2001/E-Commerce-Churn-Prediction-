```python
Return_Data = df[df["Returns"] == 1]
Not_Return_Data = df[df["Returns"] == 0]

# Calculate Return Rate by Category
Return_rate_by_Category = Return_Data["Product Category"].value_counts() / df["Product Category"].value_counts()

# Print the Return Rate by Category in bold with formatted percentages
print(f"{Fore.BLACK}{Style.BRIGHT}\nReturn Rate By Category:{Style.RESET_ALL}")

for category, rate in Return_rate_by_Category.items():
    formatted_rate = "{:.2%}".format(rate)
    print(f"{Fore.BLACK}{Style.BRIGHT}{category}: {formatted_rate}{Style.RESET_ALL}")
```

```python
payment_method_counts=df["Payment Method"].value_counts()
Total_customers=len(df)

# PRINT PAYMENT METHOD COUNTS IN BOLD
print(f"{Fore.BLACK}{Style.BRIGHT}Payment Method In Counts")
print(payment_method_counts)

# PRINT PAYMENT METHOD PERCENTAGES IN BOLD 
print(f"{Fore.BLACK}{Style.BRIGHT}\nPayment Method In Percentages")
payment_method_percentages=(payment_method_counts/Total_customers)*100
print(payment_method_percentages)
```

```python
# SORT THE DATAFRAME BASED ON "PURCHASE MONTH" AND "PURCHASE YEAR"
# IN ASCENDING ORDER, PLACING EARLIER MONTHS AND YEARS FIRST
df.sort_values(by=["Purchase Month","Purchase Year"],ascending=[True,True],inplace=True)
```

# ROC Curves & AUC For Each Classifier

The **ROC-AUC score is like a report card for a model in a binary test (yes or no)**. It checks how well the model can separate positive from negative things. **The ROC curve is a graph that shows how good the model is at making this separation. The higher the ROC-AUC score, the better the model is doing overall.**

**------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------**

```python
# COMPUTE ROC CURVES AND AUC FOR EACH CLASSIFIER
fpr_log, tpr_log, _ =metrics.roc_curve(y_test, log_pred)
fpr_rf, tpr_rf, _ =metrics.roc_curve(y_test, rf_pred)
fpr_xgb, tpr_xgb, _ =metrics.roc_curve(y_test, xgb_pred)

auc_log =metrics.auc(fpr_log, tpr_log)
auc_rf =metrics.auc(fpr_rf, tpr_rf)
auc_xgb =metrics.auc(fpr_xgb, tpr_xgb)
```

```python
# PLOT ROC CURVES
plt.figure(figsize=(8, 8), dpi=100)

plt.plot(fpr_log, tpr_log, marker='.', label=f"Logistic Regression (AUC = {auc_log:.3f})")
plt.plot(fpr_rf, tpr_rf, marker='.', label=f"Random Forest (AUC = {auc_rf:.3f})")
plt.plot(fpr_xgb, tpr_xgb, marker='.', label=f"XGBoost (AUC = {auc_xgb:.3f})")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.title("ROC Curves For Different Classifiers")
plt.show()
```
